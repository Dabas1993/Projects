## Exploratory Data Analysis

### Gender

```python
ax = sns.countplot(x = 'Sex', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Sex', hue = 'Pclass')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df1 = df.groupby (['Sex'], as_index=False)['Fare'].sum().sort_values(by='Fare', ascending=False)
sns.barplot(x='Sex', y='Fare', data = df1)


ax = sns.countplot(data = df, x = 'Sex')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Sex', hue = 'Survived')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Survived', hue = 'Pclass')

for bars in ax.containers:
    ax.bar_label(bars)
```
